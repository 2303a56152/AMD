```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(42)
n_samples = 200
data = {
    "CustomerID": np.arange(1, n_samples + 1),
    "Age": np.random.randint(18, 70, size=n_samples),
    "Annual Income": np.random.randint(20000, 120000, size=n_samples),
    "Spending Score": np.random.randint(1, 100, size=n_samples),
    "Credit Score": np.random.randint(300, 850, size=n_samples)
}
df = pd.DataFrame(data)
```

```python
df.to_csv("customer_data.csv", index=False)
```

```python
features = ["Age", "Annual Income", "Spending Score", "Credit Score"]
X = df[features]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
pca_df = pd.DataFrame(data=X_pca, columns=["PC1", "PC2"])
pca_df["CustomerID"] = df["CustomerID"]
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_df["PC1"], y=pca_df["PC2"], alpha=0.7)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Visualization of Customer Segmentation")
plt.show()
```
